To generate samples for training:
1. HACopula MATLAB toolbox (recommended, source code: https://github.com/gorecki/HACopula)
2. Marshall-Olkin sampling (implemented here)
3. Conditional method (not recommended, very slow)


```python
from phi_listing import ClaytonPhi, FrankPhi, JoePhi, GumbelPhi
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi_dict = {}
phi_dict['Clayton'] = {'Phi':ClaytonPhi,'theta':5.0}
phi_dict['Frank'] = {'Phi':FrankPhi,'theta':15.0}
phi_dict['Joe'] = {'Phi':JoePhi,'theta':3.0}
phi_dict['Gumbel'] = {'Phi':GumbelPhi,'theta':3.0}
```

```python
for key in phi_dict.keys():
    phi = phi_dict[key]['Phi'](torch.tensor(phi_dict[key]['theta']))
    s = phi.sample(2,3000).detach()
    plt.plot(s[:,0],s[:,1],'.')
    plt.axis("square")
    plt.title('%s Copula'%(key))
    plt.show()
```

```python
import pickle

key = 'Clayton'
phi = phi_dict[key]['Phi'](torch.tensor(phi_dict[key]['theta']))
s = phi.sample(2,3000).detach()

d = {'samples': s}
pickle.dump(d, open('./data/synthetic/clayton_theta5_dim2.p', 'wb'))
```
